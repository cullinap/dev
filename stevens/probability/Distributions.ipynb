```python
import math
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

## Continuous Distributions (PDF)

###### Uniform Distribution

```python
u = np.random.uniform(size=10000)
sns.set_style('ticks')
_ = sns.displot(u)
```

###### Guassian Distribution

```python
x = np.random.normal(size=10000)
_ = sns.displot(x)
_ = sns.displot(x, kde=True)
```

```python
n_experiments = 10000
heads_count = [np.random.binomial(100, 0.5) for i in range (n_experiments)]
heads, event_count = np.unique(heads_count, return_counts=True)
event_proba = event_count/n_experiments
plt.bar(heads, event_proba)
_ = plt.xlabel('Heads flips (out of 100 tosses)')
```

###### Log-normal distribution

```python
x = np.random.lognormal(size=10000) # defaults to standard normal mu=0, sigma=1
_ = sns.displot(x, kde=True)
```

##### Exponential Distribution

```python
x = np.random.exponential(scale=4, size=10000) # "scale" parameter defaults to one
_ = sns.displot(x)
```

```python
_lambda_ = 3
n = 1000
```

```python
samples = np.random.poisson(_lambda_, n); samples[0:20]
```

```python
x, x_count = np.unique(samples, return_counts=True); x
```

```python
x_count
```

```python
Px = x_count/n
```

```python
plt.bar(x, Px, color='mediumpurple')
plt.title('PMF of Poisson with lambda = {}'.format(_lambda_))
plt.xlabel('x')
_ = plt.ylabel('P(x)')
```

```python
def poisson_distribution(k, lambd):
    return (lambd ** k * np.exp(-lambd)) / np.math.factorial(k)
```

```python
poisson_distribution(2,3.5)
```
