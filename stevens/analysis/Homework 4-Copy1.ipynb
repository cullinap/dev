b) show that f is differentiable at 0 and sketch the derivative of $f(x)$ :

1) 
$
\begin{align}
 \lim_{h \to 0}\dfrac{f(x + h) + f(x)}{h} = \lim_{h \to 0}\dfrac{h^2sin(1/h)}{h} =  
 \lim_{h \to 0}hsin(1/h)
\end{align}
$

2) 

Because sin(1/x) is bounded by 1:  
$
\begin{align}
 -1<= sin(1/x) <= 1
\end{align}
$

the following is also true: 
$
\begin{align}
 -\lim_{h \to 0}|h| \leq \lim_{h \to 0} hsin(1/x) \leq \lim_{h \to 0}|h|
\end{align}
$

3) therefore by the squeeze thereom we have:
$
\begin{align}
 \lim_{h \to 0} hsin(1/x) = 0
\end{align}
$

and therefore: $f$ is differentable at 0

 R_n(x) = \dfrac{e^x}{k+1!}x^{k+1}

$

For $e^1$ and $k=6$ our error is less than $10^{-2}$

```python
def calculate_remainder(r,k):
    return (r**k/math.factorial(k))*math.exp(r)

    
if calculate_remainder(1,6) < 10**-2:
    print(calculate_remainder(1,6))
```
