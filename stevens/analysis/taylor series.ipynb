```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
first = lambda x: 1 + x 
second = lambda x: 1 + x  + (1/2*(x**2))
third = lambda x: 1 + x  + (1/2*(x**2)) + (1/6*(x**3))
fourth = lambda x: 1 + x  + (1/2*(x**2)) + (1/6*(x**3)) + (1/24*(x**4))
```

```python
x = 2
e_to_2 = 0
for i in range(5):
    e_to_2 += x**i/math.factorial(i)
    
print(e_to_2)
```

```python
x = np.linspace(-10, 10, 100000)
y = np.array([math.e**i for i in x])
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.plot(x, first(x), c='orange')
plt.plot(x, second(x), c='blue')
plt.plot(x, third(x), c='purple')
plt.plot(x, fourth(x), c='red')
_ = ax.plot(x,y)
```
