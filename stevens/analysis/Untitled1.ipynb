```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 1000)
y = lambda x: x**2

fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-1,3)
plt.ylim(-1,3)
plt.plot(x, y(x), c='red')
plt.show()
```
