You can calculate the error using: 

$
\begin{align}
 where \hspace{1mm}
 R_n(x) = \dfrac{e^x}{k+1!}x^{k+1} 
\end{align}
$

For $e^1$ and $k=6$ our error is less than $10^{-2}$

```python
def calculate_remainder(r,k):
    return (r**k/math.factorial(k))*math.exp(r)

    
if calculate_remainder(1,6) < 10**-2:
    print(calculate_remainder(1,6))
```

```python
# approximation using f(x) - R: e^x - taylor series shown

x = 1
x_to_1 = 0
error = 10**-2

for k in range(8):
    x_to_1 += x**k/math.factorial(k) 
    delta = abs(x_to_1 - math.exp(1))
    if delta < error:
        print(f'n={k+1} and delta={delta}')
        
    
```
