```python
#x = np.linspace(0,15,30)
x = [i for i in range(15)]
y = [(abs(i)*math.sin(i)) for i in x]

slope = (y[8]-y[0])/(x[8]-x[0])
gradX = x
gradY = [(slope * (i-y[0]))+y[0] for i in gradX]

fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-5,20)
plt.ylim(-10,20)
plt.vlines(x=x[0], ymin=-10, ymax=y[0], colors='r', linestyles='--')
plt.hlines(y=y[0], xmin=-5, xmax=x[0], colors='r', linestyles='--')
plt.vlines(x=x[8], ymin=-10, ymax=y[8], colors='r', linestyles='--')
plt.hlines(y=y[8], xmin=-5, xmax=x[8], colors='r', linestyles='--')

ax.plot(x,y)
ax.plot(gradX,gradY)

plt.show()
```

```python
ex3Fn = lambda x: (x**2 - (2*x) - 8)/(x - 4)
```

```python
def printLimits(limit, fn):
    for x in [limit+i for i in [-0.1, -0.001, 0.1, 0.001]]:
        print(fn(x))
```

```python
printLimits(4,ex3Fn)
```

```python
y = ex3Fn(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-5,10)
plt.ylim(-5,10)
plt.axvline(x=4, color='purple', linestyle='--')
plt.axhline(y=6, color='purple', linestyle='--')
_ = ax.plot(x,y)
```
