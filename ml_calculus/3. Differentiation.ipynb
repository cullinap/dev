```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
f = lambda x: x**2 + 2*x + 2
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_ = ax.plot(x,y)
```

```python
f(2)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, f(2))
_ = ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, f(2))
plt.scatter(5, f(5), c='orange')
_ = ax.plot(x,y)
```

```python
m = (f(5)-f(2))/(5-2); m
```

```python
b = f(5)-m*5; b
```

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, f(2))
plt.scatter(5, f(5), c='orange')
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange')
_ = ax.plot(x,y)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, f(2))
plt.scatter(2.1, f(2.1), c='orange')
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange')
_ = ax.plot(x,y)
```

```python
delta_x = 0.000001; delta_x
```

```python
x1 = -1
y1 = f(-1)
```

```python
x2 = x1 + delta_x; x2
```

```python
y2 = f(x2); y2
```

```python
m = (y2 - y1)/(x2 - x1); m
```

```python
b = x2-m*x2; b
```

```python
line_y = m*x + b
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(x1, y1)
plt.scatter(x2, y2, c='orange')
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange')
_ = ax.plot(x,y)
```

```python
diff = lambda x, delta: (f(x + delta) - f(x))/delta
```

```python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

```python
for delta in deltas:
    print(diff(2, delta))
```

```python
for delta in deltas:
    print(diff(-1, delta))
```
